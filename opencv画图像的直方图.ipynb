```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("girl.jpg")
def draw_histogram(img):
    new_img=img.ravel()#变成一个二维的向量或者list
    plt.hist(new_img,256,range=(0,256))
    plt.title("histogram")
    plt.show()
def draw_histogram_cv(img):
    colors=("blue","green","red")
    for i ,color in enumerate(colors):
        hist=cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(hist,color=color)
        plt.xlabel([0,256])
    plt.title("histogram_with_cv")
    plt.show()
        

draw_histogram(img)
draw_histogram_cv(img)
cv2.imshow("original_img",img)
if cv2.waitKey()==27:
    cv2.destroyAllWindows()
```
